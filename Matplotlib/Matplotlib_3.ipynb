# Matplotlib Ders 3

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (10,6)
```

```python
data= sns.load_dataset("penguins")
data 
```

```python
sns.histplot(data, x="flipper_length_mm") # x burada label olarak kullanÄ±lÄ±yor.
```

```python
data_male = data[data["sex"] == "Male"]["flipper_length_mm"]
data_female = data[data["sex"] == "Female"]["flipper_length_mm"]

sns.histplot(data=data_female, label = "Female", alpha=0.5, color="orange")
sns.histplot(data=data_male, label = "Male", alpha = 0.5)

plt.legend()
plt.show()
```

```python
sns.histplot(data, x="flipper_length_mm", hue= "sex")
```

## Seaborn Scatter plot

```python
sns.scatterplot(data= data, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

## Seaborn Barplot

```python
g_data = data.groupby(["species", "sex"])["body_mass_g"].mean().reset_index()
g_data
```

```python
sns.barplot(data=g_data, x="species", y="body_mass_g", hue= "sex")
plt.show()
```

## Subplot OluÅŸturma

```python
fig, axes = plt.subplots(2,2, figsize= (12,10))

sns.scatterplot(data= data, x="bill_length_mm", y="bill_depth_mm", hue="species", ax= axes[0,0])
axes[0,0].set_title("bill_length_mm vs bill_depth_mm ")
```

```python
# SORU: 
fig, axes = plt.subplots(2,2, figsize= (12,10))
sns.boxplot(data=data, x="species", y= "body_mass_g", ax=axes[0,0])
sns.histplot(data=data, x="bill_length_mm", hue="species", ax = axes[0,1])
```

## COLOR MAP EKLEME

```python
grup_data = data.groupby(["species"])["body_mass_g"].mean().reset_index()
grup_data
```

```python
sns.barplot(data=grup_data, x="species", y="body_mass_g")
```

```python
grup_d = data.groupby(["species"])["body_mass_g"].mean()

sns.barplot( x=grup_d.index, y=grup_d.values)
```

### Color Map OluÅŸturma

```python
cmap = plt.get_cmap("viridis")
norm = plt.Normalize(grup_d.min(), grup_d.max())
colors = cmap(norm(grup_d.values))
colors
```

```python
fig, axe = plt.subplots(figsize=(12,9))

sns.barplot(x=grup_d.index, y=grup_d.values, palette= colors)

axe.set_ylabel("Average body_mass_g")
cax = fig.add_axes([0.95, 0.15, 0.05, 0.67])
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
fig.colorbar(sm, cax=cax)
plt.show()
```
