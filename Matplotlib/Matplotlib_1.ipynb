## Çoklu Grafik Çizdirme ve Legend()

```python
x= np.arange(10)
y1 = x * 2
y2 = x ** 2 
plt.figure()

plt.plot(x, y1, label= "2x")
plt.plot(x, y2, label="x^2")

plt.legend()
plt.show()
```

## plt.scatter()

```python
x= np.arange(10)
y1 = x * 2

plt.figure(figsize=(6,3))
plt.scatter(x, y1, label= "2x")

plt.legend()
plt.show()
```

```python
# 2 Çeşit Grafik 
x= np.arange(10)

plt.figure(figsize=(6,3))
plt.scatter(x, y2, label= "2x", color="black")
plt.plot(x, y2, label="line", color="red")

plt.legend()
plt.show()
```

```python
# SORU
x= np.arange(100) / 10
y1= np.cos(x)
y2= np.sin(x)


plt.scatter(x, y1, label= "cos", color="black")
plt.plot(x, y2, label="sin", color="red")

plt.legend()
plt.show()
```

## plt.plot ve plt.scatter parametreleri

### plt.scatter parametreleri
`color`: renk belirtir

`label`: grafiğin etiketi

`s`: noktaların büyüklüğü

`edgecolors`: noktaları saran renk.

`alpha`: saydamlık ve opaklık

```python
x= np.arange(100) / 10
y2= np.sin(x)

plt.scatter(x, y1, label= "cos", color="black", s= 30, edgecolors="red")

plt.legend()
plt.show()
```

### plt.plot Parametreleri

`color`: çizginin rengini belirler

`linestyle`: çizgi stilini belirler

`marker`: noktalara denk gelen kesişimleri gösterir. "o" olabilir, 
"s" olabilir.

`markersize`: 

`alpha`: saydamlık - opaklık. 1=opak, 0=saydam

`label`: grafiğin etiketi
